```python
%%sql

# Joining the auditor report table and the visits table

SELECT
    auditor_report.location_id AS audit_location,
    auditor_report.true_water_source_score,
    visits.location_id AS visit_location,
    visits.record_id
FROM
    auditor_report
JOIN
    visits
ON
    auditor_report.location_id = visits.location_id
lIMIT 10;
```

```python
%%sql

# retrieve the corresponding scores from the water_quality table


SELECT
    ar.location_id,
    v.record_id,
    e.employee_name,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS employee_score,
    ar.statements
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON ar.location_id = v.location_id
JOIN
    water_quality AS wq
    ON v.record_id = wq.record_id
JOIN
    employee AS e
    ON e.assigned_employee_id = v.assigned_employee_id
WHERE
    v.visit_count = 1
    AND ar.true_water_source_score != wq.subjective_quality_score;
```

```python
%%sql


SELECT
    ar.location_id,
    v.record_id,
    e.employee_name,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS employee_score,
    ar.statements
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON ar.location_id = v.location_id
JOIN
    water_quality AS wq
    ON v.record_id = wq.record_id
JOIN
    employee AS e
    ON e.assigned_employee_id = v.assigned_employee_id
WHERE
    v.visit_count = 1
    AND ar.true_water_source_score != wq.subjective_quality_score
LIMIT 10000;
```

```python
%%sql

SELECT
    auditor_report.location_id AS audit_location,
    auditor_report.true_water_source_score AS auditor_score,
    visits.record_id,
    water_quality.subjective_quality_score AS surveyor_score,
    water_source.type_of_water_source AS survey_source,
    auditor_report.type_of_water_source AS auditor_source
FROM
    auditor_report
JOIN
    visits
ON
    auditor_report.location_id = visits.location_id
JOIN
    water_quality
ON
    visits.record_id = water_quality.record_id
JOIN
    water_source
ON
    water_source.source_id = visits.source_id
ORDER BY
    surveyor_score DESC
LIMIT 10000;
```

```python
%%sql

# employees who gave the incorrect scores

SELECT
    ar.location_id,
    v.record_id,
    e.employee_name,
    ar.true_water_source_score AS auditor_score,
    wq.subjective_quality_score AS employee_score,
    ar.statements
FROM
    auditor_report AS ar
JOIN
    visits AS v
    ON ar.location_id = v.location_id
JOIN
    water_quality AS wq
    ON v.record_id = wq.record_id
JOIN
    employee AS e
    ON e.assigned_employee_id = v.assigned_employee_id
WHERE
    v.visit_count = 1
    AND ar.true_water_source_score != wq.subjective_quality_score;
```

```python
%%sql

# prunning out employees who made more mistakes compared to the average

WITH incorrect_records AS (
    SELECT
        ar.location_id,
        v.record_id,
        e.employee_name,
        ar.true_water_source_score AS auditor_score,
        wq.subjective_quality_score AS employee_score,
        ar.statements
    FROM
        auditor_report AS ar
    JOIN
        visits AS v
        ON ar.location_id = v.location_id
    JOIN
        water_quality AS wq
        ON v.record_id = wq.record_id
    JOIN
        employee AS e
        ON e.assigned_employee_id = v.assigned_employee_id
    WHERE
        v.visit_count = 1
        AND ar.true_water_source_score != wq.subjective_quality_score
),

error_count AS (
    SELECT
        employee_name,
        COUNT(employee_name) AS number_of_mistakes
    FROM
        incorrect_records
    GROUP BY
        employee_name
),

avg_mistakes AS (
    SELECT
        AVG(number_of_mistakes) AS avg_error_count_per_empl
    FROM
        error_count
)

SELECT
    ec.employee_name,
    ec.number_of_mistakes AS error_count
FROM
    error_count AS ec
JOIN
    avg_mistakes AS avg
    ON 1=1
WHERE
    ec.number_of_mistakes > avg.avg_error_count_per_empl;
```

```python
%%sql


# finding out what the villagers on the ground said about the employees

WITH incorrect_records AS (
    SELECT
        ar.location_id,
        v.record_id,
        e.employee_name,
        ar.true_water_source_score AS auditor_score,
        wq.subjective_quality_score AS employee_score,
        ar.statements
    FROM
        auditor_report AS ar
    JOIN
        visits AS v
        ON ar.location_id = v.location_id
    JOIN
        water_quality AS wq
        ON v.record_id = wq.record_id
    JOIN
        employee AS e
        ON e.assigned_employee_id = v.assigned_employee_id
    WHERE
        v.visit_count = 1
        AND ar.true_water_source_score != wq.subjective_quality_score
),

error_count AS (
    SELECT
        employee_name,
        COUNT(employee_name) AS number_of_mistakes
    FROM
        incorrect_records
    GROUP BY
        employee_name
),

avg_mistakes AS (
    SELECT
        AVG(number_of_mistakes) AS avg_error_count_per_empl
    FROM
        error_count
)

SELECT
    ic.location_id,
    ec.employee_name,
    ic.statements
FROM
    error_count AS ec
JOIN
    avg_mistakes AS avg
    ON 1=1
JOIN
    incorrect_records AS ic
    ON ec.employee_name = ic.employee_name
WHERE
    ec.number_of_mistakes > avg.avg_error_count_per_empl;
```
